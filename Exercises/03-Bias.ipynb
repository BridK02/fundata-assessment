**Give three real-world examples of different types of cognitive bias.**

## 1. Confirmation bias 
The tendency to search for, interpret, favor, and recall information in a way that confirms or supports one's prior beliefs or values. People display this bias when they select information that supports their views, ignoring contrary information, or when they interpret ambiguous evidence as supporting their existing attitudes. The effect is strongest for desired outcomes, for emotionally charged issues, and for deeply entrenched beliefs. [Wikipedia](https://en.wikipedia.org/wiki/Confirmation_bias)


## 2. Group attribution error
This error describes our tendency to overgeneralize how a group of people will behave based on an interaction with only one person from that group (Pettigrew, 1979). For example, a negative experience with someone from a different group (e.g., a different culture, gender, religion, political party, etc.) might make us say that all members of that group share the same negative characteristics. Group attribution error forms part of the explanation for prejudice in social psychology.[REF](https://positivepsychology.com/cognitive-biases/)

## 3. Gender bias 
The tendency to assign specific behavior and characteristics to a particular gender without supporting evidence (Garb, 1997).
For example, complaints of pain are taken more seriously when made by male, rather than female, patients (Gawande, 2014); women are perceived as better caregivers than men (Anthony, 2004); specific clinical syndromes are more readily diagnosed in women than in men (Garb, 1997); and students often rate female lecturers lower than male lecturers (MacNell, Driscoll, & Hunt; 2014; Mitchell & Martin, 2018).[REF](https://positivepsychology.com/cognitive-biases/)

**Show that the difference between the standard deviation calculations is greatest for small sample sizes.**

### First I will start by showing the standard deviation for a small sample size 

```python
# Create 10000 samples of size 8 - standard deviation is 1.5.
smsamples = np.random.normal(0.0, 1.5, (10000, 8))
smsamples
```

```python
# Calculate standard deviation without correction.
smstdevs = smsamples.std(axis=1)
smstdevs
```

```python
# View a histogram - and the estimate is too small.
fig, ax = plt.subplots(figsize=(14, 9))

# Plot histogram.
plt.hist(smstdevs, bins=25)

# Draw a vertical line where the actual standard deviation is.
plt.axvline(x=1.5, color='blue')
```

The standard deviation generated without using Bessels correction in the graph is somewhere around 1.2

```python
#where as the calculated standard deviation 
x=smsamples
```

```python
np.mean(np.square(x - np.mean(x)))
```

```python
np.sqrt(np.mean(np.square(x - np.mean(x))))
```

### Now showing the standard deviation for a large sample size 

```python
# Create 10000 samples of size 100 - standard deviation is 1.5.
bigsamples = np.random.normal(0.0, 1.5, (10000, 100))
bigsamples
```

```python
# Calculate standard deviation without correction.
bigstdevs =bigsamples.std(axis=1)
bigstdevs
```

```python
# View a histogram - and the estimate is not as far off as it was for the small samples.
fig, ax = plt.subplots(figsize=(14, 9))

# Plot histogram.
plt.hist(bigstdevs, bins=30)

# Draw a vertical line where the actual standard deviation is.
plt.axvline(x=1.5, color='green')
```
